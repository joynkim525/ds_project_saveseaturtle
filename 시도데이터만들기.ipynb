```python
allt = pd.concat([t2003,t2004,t2005,t2006,t2007,t2008,t2009,t2010,t2011,t2012,t2013,t2014,t2015,t2016,t2017], sort=False)
```

```python
시도 = allt[(allt.시군구=='소계') | (allt.시군구=='합계')]
```

```python
시도 = 시도.reset_index()
```

```python
시도.시도.unique()
```

```python
시도.loc[시도[시도['시도']==' 울산 '].시도.index, '시도'] = '울산'
시도.loc[시도[시도['시도']==' 대전 '].시도.index, '시도'] = '대전'
시도.loc[시도[시도['시도']=='합 계'].시도.index, '시도'] = '전국'
```

```python
시도.drop('Unnamed: 0', axis=1)
```

```python
시도 = 시도.drop(['Unnamed: 0','재활용의류','재활용형광등','재활용윤활유', '음식물류폐기물분리배출', '음식물류폐기물분리배출', '발생량', '연탄재', '재활용폐식용유', '재활용전지류', '재활용타이어', '재활용가구류', 'index'], axis=1)
```

```python
시도.to_csv('03~17_시도쓰레기.csv', encoding='ms949')
```
